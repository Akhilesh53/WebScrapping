```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.ams.usda.gov/rules-regulations/hemp/dea-laboratories?field_lab_location_administrative_area=All&page=0'
req = requests.get(url)
soup = BeautifulSoup(req.text,"html.parser")
print(soup.prettify())
```

```python
tb = soup.find("tbody")
type(tb)
```

```python
tb
```

```python
lis = [dict()]
```

```python
import re
```

```python
bad_chars = [';',',', '.', '(', ')','-']

for t in tb.find_all("tr"):
  i=0;
  dic = dict()
  for d in t.find_all("td"):
    if i==0:
      dic["Lab Name"] = d.text
      for l in d.find_all("a",href=True):
        if(l.text):
          link = l['href']
          
          link = "https://www.ams.usda.gov/"+link
          
          #print(link)
      
      req2 = requests.get(link)
      soup2 = BeautifulSoup(req2.text,"html.parser")
      #print("prettify")
      
      lc =soup2.find("div",attrs={"class":"field__label"})
      div = lc.find_next_sibling("div")
      #print(div.text)
      dic["Lab Location"] = div.text
      
      pn = soup2.find("div",attrs={"class":"field field--name-field-phone-number field--type-string field--label-above"})
      pn_txt=pn.find("div",attrs={"class":"field__item"})
      #print(pn_txt.text)
      dic["Phone Number"] = pn_txt.text

      contact = soup2.find("div",attrs={"class":"field field--name-field-contact field--type-string field--label-above"})
      contact_txt=contact.find("div",attrs={"class":"field__item"})
      #print(contact_txt.text)
      dic["Contact"] = contact_txt.text

      email = soup2.find("div",attrs={"class":"field field--name-field-contact-email field--type-email field--label-above"})
      email_txt = email.find("div",attrs={"class":"field__item"})
      #print(email_txt.text)
      dic["Email"] = email_txt.text
    
      #print(link)
      i =(i+1)%3
    elif i==1:
      #print(d.text)
      dic["City"] = d.text

  lis.append(dic)
```

```python
print(len(lis))
```

```python
del lis[0:1]
```
