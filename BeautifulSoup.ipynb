```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://hdoa.hawaii.gov/hemp/licensees-by-county/'
```

```python
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text,"html.parser")
print(soup.prettify())
```

```python
table1 = soup.find('table',attrs={"border": '1',"style":"border-collapse: collapse; width: 91.1805%; height: 1318px;"})
table1
```

```python
table2 = soup.find('table',attrs={"border": '1',"style":"border-collapse: collapse; width: 91.1805%; height: 415px;"})
type(table2)
```

```python
table1.append(table2)
table1
```

```python
lis =[dict()]
```

```python
for tbody in table1.find_all("tbody"):
  for tr in tbody.find_all("tr"):
    i=0
    dic = dict()
    for td in tr.find_all("td"):
      if i==0:
        print("Date " + td.text)
        dic["Date"]=td.text
        i = (i+1)%4
      elif i==1:
        print("Name "+td.text)
        dic["Name"]=td.text
        i =(i+1)%4
      elif i==2:
        print("Island "+td.text)
        dic["Island"]=td.text
        i =(i+1)%4
      elif i==3:
        print("Contact "+td.text)
        dic["Contact"]=td.text
        i =(i+1)%4
    
    
    lis.append(dic)


```

```python
print(*lis,sep="\n")
```

```python
del lis[0:2]
```
